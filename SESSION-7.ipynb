# TASK - 3.

```python
num_items = int(input("enter the number of items :"))
print("the number of items is :",num_items)
total_price = 0

for i in range(num_items):
    price = float(input("enter the price of items : "))
    print("the price of items is :",price)
    total_price = total_price + price
    i = i + 1

if total_price > 2000 :
    discount = 0.1 * total_price
    print("you get 10 percent discount on your total price of all items.")

elif total_price > 1000:
    discount = 0.5 * total_price
    print("you get a 5 percent discount on your total price of all items.")
else:
    discount = 0
    print("you get no discount.")
total_bill = total_price - discount
print("So, the total price of all item is :",total_bill)
```
